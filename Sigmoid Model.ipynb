```python
import numpy as np
import matplotlib.pyplot as plt
```

### Sigmoid model - 2 Users

```python
X1 = 6
X2 = 4

# Sigmoid parameters
sigmoid_scale = 0.05
sigmoid_shift = 5

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-sigmoid_scale * (x - sigmoid_shift)))

initial_C = 10
iterations = 50
X1_values = [X1]
X2_values = [X2]
for i in range(iterations):
    if X1 + X2 < initial_C:
        X1 = X1 + alpha1 * sigmoid(X1)
        X2 = X2 + alpha2 * sigmoid(X2)
    else:
        X1 = X1 * beta1
        X2 = X2 * beta2
    X1_values.append(X1)
    X2_values.append(X2)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(X2_values, X1_values, label='X2 vs X1', color='blue')  # Plotting X2 vs X1
plt.xlabel('X1 (Window Size)')
plt.ylabel('X2 (Window Size)')
plt.title('Sigmoid Algo - 2 Users')

# Plotting the line passing through the origin with gradient 1
max_value = max(max(X1_values), max(X2_values))
plt.plot([0, max_value], [0, max_value], linestyle='--', color='green', label='Fairness Line')


eff_X = np.linspace(0, max(X1_values), 100)  # Example X1 values
eff_y = initial_C-eff_X
plt.plot(eff_X, eff_y, linestyle='--', color='red', label='Efficiency Line')

last_X1 = X1_values[-1]
last_X2 = X2_values[-1]
plt.plot(last_X2, last_X1, marker='X',color='red', label='Final Value')


plt.legend()
plt.grid(True)
plt.show()

iterations_axis = list(range(iterations + 1))  # Adding 1 to include the initial values
initial_C_line = [initial_C] * (iterations + 1)  # Straight line for initial_C

plt.figure(figsize=(12, 6))

plt.plot(iterations_axis, X1_values, label='X1')
plt.plot(iterations_axis, X2_values, label='X2')

intersection = initial_C/2
plt.plot(intersection, intersection, marker='X',color='green', label='Optimal Point')


plt.xlabel('Iterations')
plt.ylabel('Values')
plt.title('Sigmoid Model - 2 Users')
plt.legend()
plt.grid(True)
plt.show()
```

### Sigmoid Model 5 Users

```python
import numpy as np
import matplotlib.pyplot as plt

alpha1 = 1  # Fixed additive increase parameter
beta1 = 0.5
alpha2 = 1
beta2 = 0.5
alpha3 = 1
beta3 = 0.5
alpha4 = 1
beta4 = 0.5
alpha5 = 1
beta5 = 0.5

X1 = 5
X2 = 3
X3 = 2
X4 = 5
X5 = 5


# Sigmoid parameters
sigmoid_scale = 0.05
sigmoid_shift = 5

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-sigmoid_scale * (x - sigmoid_shift)))

initial_C = 20
iterations = 50

# Lists to store values for plotting
X1_values = [X1]
X2_values = [X2]
X3_values = [X3]
X4_values = [X4]
X5_values = [X5]

# AIMD algorithm
for i in range(iterations):
    if X1 + X2 + X3 + X4 + X5 < initial_C:
        X1 = X1 + alpha1 * sigmoid(X1)
        X2 = X2 + alpha2 * sigmoid(X2)
        X3 = X3 + alpha3 * sigmoid(X3)
        X4 = X4 + alpha4 * sigmoid(X4)
        X5 = X5 + alpha5 * sigmoid(X5)
    else:
        X1 = X1 * beta1
        X2 = X2 * beta2
        X3 = X3 * beta3
        X4 = X4 * beta4
        X5 = X5 * beta5
        
    X1_values.append(X1)
    X2_values.append(X2)
    X3_values.append(X3)
    X4_values.append(X4)
    X5_values.append(X5)

# Plotting
iterations_axis = range(iterations + 1)  # +1 to include initial values

plt.figure(figsize=(12, 6))

plt.plot(iterations_axis, X1_values, label='X1')
plt.plot(iterations_axis, X2_values, label='X2')
plt.plot(iterations_axis, X3_values, label='X3')
plt.plot(iterations_axis, X4_values, label='X4')
plt.plot(iterations_axis, X5_values, label='X5')

plt.xlabel('Iterations')
plt.ylabel('Values')
plt.title('Sigmoid Algorithm - 5 Users')
plt.legend()
plt.grid(True)
plt.show()
```
